So clearly we need to filter out the low frequency names. So let's explore the dataset features to see if we can get rid of the above names.

```python
def get_names_set_at_nobs(min_nobs=2000):
    mask = df["nobs"] > min_nobs
    filtered_name_set = set(df[mask].name)
    return filtered_name_set

print(get_names_set_at_nobs())
```
